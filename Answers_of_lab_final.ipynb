```python
# Upload the image(DIP_Lab_task_01.png) from drive with extracting details like: image dimension, frequency distribution of intensities, range. [5]

import cv2
import matplotlib.pyplot as plt
import numpy as np

image_path = '/content/drive/MyDrive/Assignment/Assignment/DIP_Lab_task_01.png'
image = cv2.imread(image_path)
print(f"Image Dimensions: {image.shape}")

histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.title("Frequency Distribution")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(histogram)
plt.show()

pixel_range = image.max() - image.min() + 1
print(f"Pixel Intensity Range: {pixel_range}")
```

```python
# Convert the RGB image(DIP_Lab_task_02.jpg) to a gray scale image and compare a sub-region of the image by intensity values(5*5 dimension would be enough for comparison). [10]

from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('/content/drive/MyDrive/Assignment/Assignment/DIP_Lab_task_02.jpg')

# Convert RGB to grayscale
r, g, b = img[..., 0], img[..., 1], img[..., 2]
gray_img = (r * 0.25 + g * 0.50 + b * 0.25).astype(np.uint8)

# Plot the images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(img)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(gray_img, cmap='gray')
ax[1].set_title('Grayscale Image')
ax[1].axis('off')

plt.show()
print(ax[1].get_xlim(), ax[1].get_ylim())

```

```python
# Apply Image(DIP_Lab_task_03.png) smoothing with appropriate kernel applied. Identify the noise(if any) before processing.

from skimage import io, filters
import matplotlib.pyplot as plt

img = io.imread('/content/drive/MyDrive/Assignment/Assignment/DIP_Lab_task_03.png', as_gray=True)

# Gaussian smoothing
smoothed_img = filters.gaussian(img, sigma=1)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(smoothed_img, cmap='gray')
plt.title('Smoothed Image')
plt.axis('off')

plt.show()
```

```python
# task4-- Implement an appropriate technique to reduce the brightness of the provided image (DIP_Lab_task_04.jpg) effectively.
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

sample_img = cv2.imread('/content/drive/MyDrive/Assignment/Assignment/DIP_Lab_task_04.jpg')

normalizedImg = sample_img / 255.0

gamma = 2.1

image_gamma_corrected = np.power(normalizedImg, gamma)

image_gamma_corrected = np.uint8(image_gamma_corrected * 255)

cv2.imwrite('reduced_brightness_image.jpg',image_gamma_corrected)

cv2_imshow(image_gamma_corrected)
```
